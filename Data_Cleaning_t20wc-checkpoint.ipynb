```python
import pandas as pd
import json
import numpy as np
```

```python
with open('t20_wc_match_results_20230704_164358.json') as f:
  data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match
```

```python
df_match.rename({'scorecard' : 'match_id'} , axis = 1 , inplace = True)
```

```python
df_match.head(5)
```

```python
match_ids_dict = {}
for index , row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
    
match_ids_dict 
```

```python
df_match.to_csv('t20_wc_summary.csv' , index = False)
```

BATTING SUMMARY

```python
with open('t20_batting_20230704_163816.json') as bt:
     data = json.load(bt)
all_records = []
for rec in data:
        all_records.extend(rec['battingSummary'])
all_records

df_batting = pd.DataFrame(all_records)

```

```python
df_batting['batsmanName'] = pd.Series(df_batting['batsmanName'].str.decode("utf-8", "ignore"))

```

```python
df_batting.head(11)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'not_out')
df_batting.head(11)
```

```python
df_batting['dismissal'] = df_batting['dismissal'].replace(np.nan , " ")
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 'out' if len(str(x)) > 1 else 'not_out')
```

```python
df_batting.head(11)
```

```python
df_batting.head(11)
```

```python
df_batting["match_id"] = df_batting['match'].map(match_ids_dict)
df_batting.head(40)
```

```python
df_batting
```

```python
df_batting.to_csv('t20_battin_summary.csv' , index = False)
```
